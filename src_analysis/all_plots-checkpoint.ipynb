```python
import pandas as pd
import numpy as np
from sklearn.externals import joblib
```

```python
df = pd.read_csv(r'C:\\Users\\91817\\Desktop\\AQI-Predictor\\src_analysis\\merged_new.csv')
df.head(1)
```

```python
col_list = ["PM10", "D-1 PM10", "Air Temperature", "Horizontal Visibility", "Dew Point Temperature", "Wind Speed", "Relative Humidity"]
target_y = ['PM10']
```

```python
X = df[col_list][:-1]
y = df[target_y][1:]
```

```python
calc_test_size = round(X.shape[0]*(1 - 0.2))
x_train = X.iloc[:calc_test_size, :]
x_test = X.iloc[calc_test_size:, :]
y_train = y.iloc[:calc_test_size, :]
y_test = y.iloc[calc_test_size:, :]
xx_train = np.array(x_train)
yy_train = np.array(y_train)
xx_test = np.array(x_test)
yy_test = np.array(y_test)
```

```python
pm10_svr_obj = joblib.load(r'C:\\Users\\91817\\Desktop\\AQI-Predictor\\saved_models\\pm10\\pm10_svr_predictor.joblib')
pm10_svr_xscaler = joblib.load(r'C:\\Users\\91817\Desktop\\AQI-Predictor\\saved_scalers\\pm10\\pm10_svr_xscaler.save')
pm10_svr_yscaler = joblib.load(r'C:\\Users\\91817\Desktop\\AQI-Predictor\\saved_scalers\\pm10\\pm10_svr_yscaler.save')
```

```python
X = pm10_svr_xscaler.transform(xx_train)
prediction = pm10_svr_obj.predict(X)
predicted_result = pm10_svr_yscaler.inverse_transform(prediction)
```

```python
import matplotlib.pyplot as plt
x = np.arange(len(yy_train))
```

```python
plt.figure(figsize = (10, 10))
plt.style.use('dark_background')
plt.scatter(x, yy_train, label = "actual", c = "b", s = 100)
plt.scatter(x, predicted_result, label = "prediction", c = "skyblue", s = 100)
plt.xlabel("Days", fontsize = 13)
plt.ylabel("SO2", fontsize = 13)
plt.grid()
plt.xticks(size = 10)
plt.yticks(size = 10)
plt.tight_layout()
plt.title("Support vector : SO2 scatter plot", fontsize = 15)
plt.legend(fontsize = 12)
plt.show()
```
